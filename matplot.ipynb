```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[10,15,13,17,20]

# create line plot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Line Plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,15,13,17,20]

# create customized line plot 
plt.plot(x,y,color="purple",linestyle='--')
```

```python
x=[1,2,3,4,5]
y=[10,15,13,17,20]

# create line plot
plt.plot(x,y,color="purple",linestyle='--',marker='o',linewidth=4,markersize=9)
plt.grid(True)
```
